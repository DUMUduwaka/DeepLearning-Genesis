### Warmup 01 - Sleep in

```python
def sleep_in(weekday, vacation):
    if not weekday or vacation:
        return True
    else:
        return False
```

### Warmup 01 - Near Hundred

```python
def near_hundred(n):
    if abs(100-n)<=10 or abs(200-n)<=10:
        return True
    else:
        return False
```

### Warmup 01 - Sum Double

```python
def sum_double(a,b):
    if a==b:
        return (a+b)*2
    else:
        return a+b
```

### Warmup 02 - Array Front 9

```python
def array_front9(num):
    end= len(num)
    if end > 4:
        end = 4
    
    for i in range(end):
        if num[i]==9:
            return True
    return False
```

### Warmup 02 - Array 123

```python
def array123(num):
    for i in range(len(num)-2):
        if num[i]==1 and num[i+1]==2 and num[i+2]==3:
            return True
    return False
```

### Striing 1 - make abba

```python
def make_abba(a,b):
    return a + b + b + a
```

### String 1 - Combo String

```python
def combo_string(a,b):
    if len(a)<len(b):
        return a+b+a
    else:
        return b+a+b
```

### Stirng 01 - Non Start

```python
def non_start(a,b):
    return a[1:]+b[1:]
```

### Logic 01 - Love 6

```python
def love6(a,b):
    if a==6 or b==6 or abs(a-b)==6 or a+b==6:
        return True
    return False
```

### Logic 01 - Squirrel Play

```python
def squirrel_play(temp, is_summer):
    if is_summer and 60 <= temp <= 100:
        return True
    elif not is_summer and 60 <= temp <= 90:
        return True
    
    return False
```

### Logic 01 - Caught Speeding 

```python
def caught_speeding(speed, is_birthday):
    if is_birthday:
        if speed<= 65:
            return 0
        elif 66<= speed <= 85:
            return 1
        elif speed >= 86:
            return 2
    
    if not is_birthday:
        if speed<= 60:
            return 0
        elif 61<= speed <= 80:
            return 1
        elif speed >= 81:
            return 2
```

### Logic 02 - Make Bricks

```python
def make_bricks(small, big, goal):
    if goal > (big*5 + small):
        return False
    if (goal % 5) > small:
        return False
    return True
```
